```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from node2vec import Node2Vec
import torch
import torch.nn.functional as F
from torch_geometric.data import Data, DataLoader
from torch_geometric.nn import GCNConv

# Function to preprocess the dataset and create graph
def preprocess_dataset(file_path):
    # Read the dataset
    data = pd.read_csv(file_path)

    # Create a NetworkX graph
    G = nx.Graph()

    # Add nodes to the graph
    for idx, row in data.iterrows():
        G.add_node(idx, protocol_type=row['protocol_type'], service=row['service'])

    # Optionally, add edges based on certain conditions
    for idx1, row1 in data.iterrows():
        for idx2, row2 in data.iterrows():
            if idx1 != idx2:
                if row1['protocol_type'] == row2['protocol_type']:
                    G.add_edge(idx1, idx2)

    # Visualize the graph (optional)
    plt.figure(figsize=(10, 6))
    pos = nx.spring_layout(G, k=0.5)
    nx.draw(G, pos, with_labels=True, width=0.4, node_color='lightblue', node_size=400)
    plt.show()

    # Initialize Node2Vec
    node2vec = Node2Vec(G, dimensions=64, walk_length=30, num_walks=200, workers=1)  # Use only one worker to avoid pickling issues

    # Train Node2Vec model
    model = node2vec.fit(window=10, min_count=1, batch_words=4)

    # Get node embeddings
    node_embeddings = {node: model.wv[node] for node in G.nodes()}

    # Normalize node embeddings
    scaler = StandardScaler()
    node_embeddings_normalized = scaler.fit_transform(list(node_embeddings.values()))

    # Display normalized node embeddings
    print("Normalized Node Embeddings:")
    print(node_embeddings_normalized)

    # Create edge index
    edge_index = torch.tensor(list(G.edges)).t().contiguous()

    # Create node features tensor
    x = torch.tensor(node_embeddings_normalized, dtype=torch.float)

    # Retrieve and map labels
    data['class'] = data['class'].map({'normal': 0, 'anomaly': 1})
    labels = torch.tensor(data['class'].values, dtype=torch.long)

    return Data(x=x, edge_index=edge_index, y=labels)

# Example usage
train_file_path = 'your_dataset.csv'
test_file_path = 'your_dataset1.csv'
train_data = preprocess_dataset(train_file_path)
test_data = preprocess_dataset(test_file_path)

# Define the GCN model
class GCN(torch.nn.Module):
    def __init__(self, num_node_features, num_classes):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(num_node_features, 32)  # Increase number of neurons
        self.conv2 = GCNConv(32, num_classes)  # Adjust output size

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, training=self.training)  # Apply dropout for regularization
        x = self.conv2(x, edge_index)
        return F.log_softmax(x, dim=1)

# Load the data
train_loader = DataLoader([train_data], batch_size=1, shuffle=True)
test_loader = DataLoader([test_data], batch_size=1, shuffle=False)

# Train and evaluate the model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = GCN(num_node_features=train_data.num_features, num_classes=2).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)

def train():
    model.train()
    for batch in train_loader:
        batch = batch.to(device)
        optimizer.zero_grad()
        out = model(batch)
        loss = F.nll_loss(out, batch.y)
        loss.backward()
        optimizer.step()

def test(loader):
    model.eval()
    correct = 0
    total_samples = 0
    for batch in loader:
        batch = batch.to(device)
        out = model(batch)
        pred = out.max(dim=1)[1]
        correct += pred.eq(batch.y).sum().item()
        total_samples += batch.y.size(0)
    return correct / total_samples if total_samples > 0 else 0.0

# Training loop
for epoch in range(50):
    train()
    train_acc = test(train_loader)
    print(f'Epoch: {epoch:03d}, Train Acc: {train_acc:.4f}')

test_acc = test(test_loader)
print(f'Test Accuracy: {test_acc:.4f}')
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from node2vec import Node2Vec
import torch
import torch.nn.functional as F
from torch_geometric.data import Data, DataLoader
from torch_geometric.nn import GCNConv, GATConv, global_mean_pool
from torch_geometric.utils import k_hop_subgraph

# Function to preprocess the dataset and create graph
def preprocess_dataset(file_path):
    # Read the dataset
    data = pd.read_csv(file_path)

    # Create a NetworkX graph
    G = nx.Graph()

    # Add nodes to the graph
    for idx, row in data.iterrows():
        G.add_node(idx, protocol_type=row['protocol_type'], service=row['service'])

    # Optionally, add edges based on certain conditions
    for idx1, row1 in data.iterrows():
        for idx2, row2 in data.iterrows():
            if idx1 != idx2:
                if row1['protocol_type'] == row2['protocol_type']:
                    G.add_edge(idx1, idx2)

    # Visualize the graph (optional)
    plt.figure(figsize=(10, 6))
    pos = nx.spring_layout(G, k=0.5)
    nx.draw(G, pos, with_labels=True, width=0.4, node_color='lightblue', node_size=400)
    plt.show()

    # Initialize Node2Vec
    node2vec = Node2Vec(G, dimensions=64, walk_length=30, num_walks=200, workers=1)  # Use only one worker to avoid pickling issues

    # Train Node2Vec model
    model = node2vec.fit(window=10, min_count=1, batch_words=4)

    # Get node embeddings
    node_embeddings = {node: model.wv[node] for node in G.nodes()}

    # Normalize node embeddings
    scaler = StandardScaler()
    node_embeddings_normalized = scaler.fit_transform(list(node_embeddings.values()))

    # Display normalized node embeddings
    print("Normalized Node Embeddings:")
    print(node_embeddings_normalized)

    # Create edge index
    edge_index = torch.tensor(list(G.edges)).t().contiguous()

    # Create node features tensor
    x = torch.tensor(node_embeddings_normalized, dtype=torch.float)

    # Retrieve and map labels
    data['class'] = data['class'].map({'normal': 0, 'anomaly': 1})
    labels = torch.tensor(data['class'].values, dtype=torch.long)

    return Data(x=x, edge_index=edge_index, y=labels)

# Example usage
train_file_path = 'your_dataset.csv'
test_file_path = 'your_dataset1.csv'
train_data = preprocess_dataset(train_file_path)
test_data = preprocess_dataset(test_file_path)

# Define the GAT model
class GAT(torch.nn.Module):
    def __init__(self, num_node_features, num_classes):
        super(GAT, self).__init__()
        self.conv1 = GATConv(num_node_features, 32, heads=4, concat=True)  # Multi-head attention
        self.conv2 = GATConv(32 * 4, num_classes, heads=1, concat=True)  # Output heads=1 and concat=False

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, training=self.training, p=0.6)  # Apply dropout for regularization
        x = self.conv2(x, edge_index)
        return F.log_softmax(x, dim=1)

# Load the data
train_loader = DataLoader([train_data], batch_size=1, shuffle=True)
test_loader = DataLoader([test_data], batch_size=1, shuffle=False)

# Train and evaluate the model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = GAT(num_node_features=train_data.num_features, num_classes=2).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.005, weight_decay=5e-4)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=20, gamma=0.5)  # Add learning rate scheduler

def train():
    model.train()
    for batch in train_loader:
        batch = batch.to(device)
        optimizer.zero_grad()
        out = model(batch)
        loss = F.nll_loss(out, batch.y)
        loss.backward()
        optimizer.step()

def test(loader):
    model.eval()
    correct = 0
    total_samples = 0
    for batch in loader:
        batch = batch.to(device)
        out = model(batch)
        pred = out.max(dim=1)[1]
        correct += pred.eq(batch.y).sum().item()
        total_samples += batch.y.size(0)
    return correct / total_samples if total_samples > 0 else 0.0

# Training loop
for epoch in range(50):
    train()
    train_acc = test(train_loader)
    print(f'Epoch: {epoch:03d}, Train Acc: {train_acc:.4f}')
    scheduler.step()  # Step the learning rate scheduler

test_acc = test(test_loader)
print(f'Test Accuracy: {test_acc:.4f}')
```
